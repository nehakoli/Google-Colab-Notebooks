# ðŸŸ¢ K-Means Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

Generate synthetic data with 4 clusters

```python
# generate synthetic data using the make_blobs function from scikit-learn,
# which creates clusters of points with specified centers and standard deviation.
# cluster_std=0.60 (it controls how spread out the points within each cluster are)

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60)
```

```python
# Visualize synthetic data
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Data')
plt.show()
```

Instantiate KMeans with the number of clusters (K) you want

```python
kmeans = KMeans(n_clusters=4)
```

```python
# Fit KMeans to the data
kmeans.fit(X)
```

Get the cluster centroids

```python
centroids = kmeans.cluster_centers_
```

Get labels for each point in X (cluster assignments)

```python
labels = kmeans.labels_
```

Visualize the clusters and centroids

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', c='red', s=200, alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.show()
```

# ðŸŸ£ Agglomerative (Hierarchical) Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

Generate synthetic data again with 4 clusters

```python
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

Perform hierarchical clustering using Ward linkage

```python
agg_clustering = AgglomerativeClustering(n_clusters=4, linkage='ward')
agg_clustering.fit(X)
```

Visualize the clustered data

```python
plt.scatter(X[:, 0], X[:, 1], c=agg_clustering.labels_, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Agglomerative Hierarchical Clustering')
plt.show()
```
